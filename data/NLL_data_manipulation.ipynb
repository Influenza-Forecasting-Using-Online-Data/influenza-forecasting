Calculate Variance needed for Negative Log Likelihood. 
- starting from 2009, we use data points from the same week from past years to calculate the mean and the variance. 
- for 2009, it uses past 5 years of data, from the same week
- for 2010, it uses past 6 years of data, again from the same week. 


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datetime_data_frame.csv")
```

```python
#calculate mean and variance for weeks from 2009
df.rename(columns = {"Unnamed: 0": "index"}, inplace= True)
df_from_2009 = df[df["year"] >= 2009] #put a condition in the brackets
idx_list = list(df_from_2009["index"])

df["Mean"] = np.nan
df["Variance"] = np.nan
# idx_list
for idx in idx_list:
    curr_row = df[idx: idx+1]
    current_week = int(curr_row["week"])

    prev_rows = df[:idx] #currentRow+1, to include current row in the df. 
    prev_rows_week_n = prev_rows[prev_rows["week"]==current_week] #df of week n
#     if current_week == 1: 
#         print(prev_rows_week_n)
    if len(prev_rows_week_n)<=1: 
        df_mean = np.nan
        df_variance = np.nan #for edge case (ie week 53)
    else: 
        df_mean= prev_rows_week_n['Disease Rate'].mean()
        df_variance = prev_rows_week_n['Disease Rate'].var()
    df.at[idx, "Mean"] = df_mean
    df.at[idx, "Variance"] = df_variance

df
#mean and variance, verified through calculator.
#for 2009. we use data points from 2004 to 2009 inclusive. to calculate the mean & variance.
```

```python
df.to_csv("datetime_data_frame_mean_variance.csv", index= False, header= True)
```
